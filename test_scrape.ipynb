```python
def date_to_timestamp(date):
    return int(time.mktime(datetime.strptime(date, "%Y-%m-%d").timetuple()))
```

```python
start_date = date_to_timestamp('2000-01-01')
end_date = date_to_timestamp('2024-01-01')
```

```python
driver = webdriver.Edge()

all_data = []

for _, row in nasdaq_tickers_df.iterrows():
    ticker = row['Ticker']
    company_name = row['Name']
    url = f"https://finance.yahoo.com/quote/{ticker}/history?period1={start_date}&period2={end_date}&interval=1d&filter=history&frequency=1d"
    driver.get(url)
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    table = soup.find('table')

    headers = [header.text for header in table.find_all('th')]

    rows = []
    for row in table.find_all('tr')[1:]:
        cols = row.find_all('td')
        if len(cols) > 1:
            rows.append([col.text for col in cols])

    df_each_ticker = pd.DataFrame(rows, columns=headers)
    df_each_ticker['Ticker'] = ticker
    df_each_ticker['Company Name'] = company_name
    all_data.append(df_each_ticker)

driver.quit()

df = pd.concat(all_data, ignore_index=True)

df.to_csv('scraped_stock_data.csv', index=False)
```
